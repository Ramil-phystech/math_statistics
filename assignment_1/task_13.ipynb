```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
alpha = 0.05
n = 139
m = 1000
```

```python
quantile_alpha_2 = stats.f.ppf(alpha / 2, n - 1, m - 1)
print(f"Квантиль порядка alpha/2:\n{quantile_alpha_2}")
```

```python
quantile_1_minus_alpha_2 = stats.f.ppf(1 - alpha / 2, n - 1, m - 1)
print(f"Квантиль порядка 1 - alpha/2:\n{quantile_1_minus_alpha_2}")
```

```python
Sx_length, Sy_length = 5.722, 6.161
Sx_width, Sy_width = 4.612, 5.055
```

```python
delta_length = Sx_length ** 2 / Sy_length ** 2
print(f"Значение delta для длины:\n{delta_length}")
```

```python
delta_width = Sx_width ** 2 / Sy_width ** 2
print(f"Значение delta для ширины:\n{delta_width}")
```

### => H0 не отвергаем

```python
_, axs = plt.subplots(1, 1, figsize=(10, 6))

axs.set_title("Мощность критерия")
axs.set_xlabel("a (отношение стандартных отклонений)")
axs.set_ylabel("W(a)")

a = np.linspace(0.5, 2, 100)

power = (1 - stats.f.cdf(quantile_1_minus_alpha_2, n - 1, m - 1, scale=a**2) + 
         stats.f.cdf(quantile_alpha_2, n - 1, m - 1, scale=a**2))

axs.plot(a, power, color="green", label="Мощность W(a)")

axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(10))

plt.grid(which='minor', linestyle='--', linewidth=0.2)
plt.grid(which='major', linewidth=0.4)

plt.legend()
plt.show()
```
